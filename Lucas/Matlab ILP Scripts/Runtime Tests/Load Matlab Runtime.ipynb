```python
def parseData(filename):
    mat = loadmat(filename)
    data = mat['tests'][0]
    ILP_times = {}
    for d in data:
        d = d[0]
        As = list(ILP_times.keys())
        k = d[0][0][0]
        a = d[1][0][0]
        R = d[2][0]
        R = R.split(',')
        isResolving = d[3][0][0]
        times = d[4][0]
        if a in As:
            ks = list(ILP_times[a].keys())
            for time in times:
                if k in ks:
                    ILP_times[a][k].append((R,isResolving,time))
                else:
                    ILP_times[a][k] = [(R,isResolving,time)]
        else:
            ILP_times[a] = {}
            for time in times:
                if k in list(ILP_times[a].keys()):
                    ILP_times[a][k].append((R,isResolving,time))
                else:
                    ILP_times[a][k] = [(R,isResolving,time)]
    return ILP_times
```

```python
filename = 'matlab_runtime.mat'
ILP_times = parseData(filename)
```

```python
with open('ILP_times_all.dict','wb') as f:
    pickle.dump(ILP_times,f)
```
